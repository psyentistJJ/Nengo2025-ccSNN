```python
def custom_agg(series):
    #somehow, some odors are duplicate
    # merge those odors: if for a receptor, only one row is non-zero, take that value, otherwise take the mean of the 2

    non_zeros = series[series != 0]
    if len(non_zeros) == 0:
        # all zeros, return 0
        return 0
    elif len(non_zeros) == 1:
        # only one non-zero value
        return non_zeros.iloc[0]
    else:
        # multiple non-zero values: return their mean
        return non_zeros.mean()


DoOR_selected = DoOR_selected.groupby(DoOR_selected.index).agg(custom_agg)

N_odors=DoOR_selected.shape[0]

odor_ORN_array = np.array(DoOR_selected)
```

```python
%%cython -a
from libc.stdlib cimport rand, RAND_MAX, srand
from libc.time cimport time


def single_odor(ORN_codex_2_door,door_cols,array,map_names_r,list input_neurons, int odor_idx,float tmax, float stim_start, float stim_end,float x0,float x1):
    '''
    -> adapted from Nanami et al. (2024)
    args:
        array: odor x receptor
        N_OR: number of receptors
        odor_idx: index of current odor (response to this opdor is simulated for all neurons)
        tmax: duration of 'recording' in s
        stim_start: stimulus (odor) onset
        stim_end: stimulus offset
        x0: firing freq without stim
        x1: 'concentration' of odor
    '''
    d0=[[]]
    responding_idxs = []
    for t in range(int(tmax*1000)):
        if stim_start<=t and t < stim_start+stim_end:
            for n, neuron in enumerate(input_neurons):
                #for every neuron sample acc to receptor it belongs to -> sample data for each neuron sepoarately (otherwise, all neurons from one class spike only at the same time)
                #i = door_cols.index(ORN_codex_2_door[neuron])
                #i is index of receptor type, not single neurons
                i_s=[]
                for receptor in ORN_codex_2_door[neuron]: #some neurons are listed ofr multiple receptors (?)
                    i_s.append(door_cols.index(map_names_r[receptor]))


                r0=0
                for i in i_s:
                    r0+=array[odor_idx][i]*x0*x1
                if r0<0:
                    r0=0
                elif r0>0:
                    if t==0:
                        responding_idxs.append(n)
                if (rand()+1.0)/(RAND_MAX+2.0)<(r0+8)*0.001:
                    l=[]
                    l.append(t*0.001)
                    l.append(n)
                    d0.append(l)
                    
                
        else:
            for n, neuron in enumerate(input_neurons):
                if (rand()+1.0)/(RAND_MAX+2.0)<8*0.001:
                    l=[]
                    l.append(t*0.001)
                    l.append(n)
                    d0.append(l)

    return d0, responding_idxs
```
